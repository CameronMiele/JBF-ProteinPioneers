### Extracting protein names from Expasy output

```python
import pandas as pd
import os
```

```python
#Creates a file path to the combined protein data file
protein_file = os.path.join('data', 'Combined.35315.txt')
print(protein_file)
```

```python
#Creates a dataframe from the file
protein_df = pd.read_csv(protein_file, sep = '\t')
protein_df
```

```python
#Creates a data frame of only each protein's description
df2 = protein_df['Description']
df2
```

```python
#Creates the variable protein_name for each entry in the database as a copy of the Description entry
protein_name = []
for line in protein_df['Description']:
    protein_name = protein_df['Description']
    print(protein_name)
```

```python
#Creates a Name entry for each protein in the data base by removing elements from the Description string
protein_df['Name'] = protein_df['Description'].str.extract(r'(?<=\=)(.*?)(?=\;)')
protein_df
```
