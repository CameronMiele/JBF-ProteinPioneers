```python
from Bio import SeqIO
from Bio.SeqUtils.IsoelectricPoint import IsoelectricPoint as IP
from Bio.SeqUtils.ProtParam import ProteinAnalysis as PA
import os
from collections import defaultdict
import pandas as pd
```

```python
#Generate a file path for the correct .fasta file

filename = 'Streptococcus_anginosus.faa'
#filename = input("Enter File Name")
print(filename)
datafile = os.path.join('data', filename)
print(datafile)
```

```python
#Create a subset of the original .fasta with some attribute

short_seqs = []
neg_at_pH7 = []
with open(datafile) as protfile:
    for record in SeqIO.parse(protfile,"fasta"):
        prot_params = PA(str(record.seq))
        if len(record.seq) < 300:
            short_seqs.append(record)
        if prot_params.charge_at_pH(7) < 0:
            neg_at_pH7.append(record)
SeqIO.write(short_seqs,os.path.join('data','Modified',input('Enter file name')),"fasta")
SeqIO.write(neg_at_pH7,os.path.join('data','Modified',input('Enter file name')),"fasta")
```

```python
#Converting modified fasta into a pd dataframe to show filter at work, preset file name used for simplicity

data = defaultdict(list)
datalen = []
protpI = []
protMW = []
prot_charge_at_pH = []
with open('data/Modified/yersinia_pestis_short_sequences.fasta') as protfile:
  for record in SeqIO.parse(protfile,"fasta"):
    sequence = str(record.seq)
    length = len(sequence)
    data['sequence'].append(sequence)
    datalen.append(length)
    protparams = PA(sequence)
    protpI.append(protparams.isoelectric_point())
    protMW.append(protparams.molecular_weight())
    prot_charge_at_pH.append(protparams.charge_at_pH(7))

df = pd.DataFrame.from_dict(data)

df['length'] = datalen

df['IEP'] = protpI

df['MW'] = protMW

df['charge at pH 7'] = prot_charge_at_pH

df.sort_values(by=["length"])
```
